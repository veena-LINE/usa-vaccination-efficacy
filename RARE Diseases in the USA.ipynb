<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

<br>Data is available as a single compressed file (easy to maintain) enveloping several files, one for each disease. A brief look at the data files reveals matching column names on `week`, `state` and `state_name` in all the files.

To further investigate the need for data cleansing, let's first read all the files into separate dataframes without the need to uncompress. Here, I will not be reading the 2-letter `state` codes as state names are available in `state_name` column which I will be renaming to `State`.
</span>
<br><br>

```python
{
    "tags": [
        "hide-input",
    ]
}

import zipfile
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

pd.plotting.register_matplotlib_converters()
%matplotlib inline

from warnings import filterwarnings
filterwarnings("ignore")

# CUSTOM COLORS
colors_line = ["#DC143C", "#008B8B", "#8B008B", "#C71585", "#191990", "#FF4500", "#808000"]
colors_bar = ["#E4F132", "#9A0794", "#FC600A", "#1489B8", "#C21460", "#B2D732", "#551130"]
```

```python
{
    "tags": [
        "hide-input",
    ]
}

diseases_dataset = 'usa_contagiousdiseases.zip'
diseases = {}

with zipfile.ZipFile(diseases_dataset, mode='r') as z:
    for csv in [csvs for csvs in z.namelist() if csvs.endswith('.csv')]:
        diseases[csv.split('.')[0]] = (
            pd.read_csv(z.open(csv, mode='r')
                        , usecols=lambda col: col != 'state'
                        , converters={'week': str}
                       ).rename(columns={'week': 'Week'
                                         , 'disease': 'Disease'
                                         , 'state_name': 'State'
                                         , 'cases': 'Cases'
                                         , 'incidence_per_capita': 'Incidence/100,000'}
                               )
        )

print(': Initial data on hand, by disease :', end='\n\n')
for disease, data in diseases.items():
    print(disease.ljust(15, ' '), str(data.shape[0]).rjust(8, ' '), 'rows of data')
```

<h4 id="duplicate-data"> |2.| Duplicate Data </h4>

</span>
<br>
<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

Data sourced from Kaggle is an aggregated data-set at the state level. This means that there should a single entry per state-time period. If there had been any duplicated entries during aggregation, we don't want to count that twice or more. Let's find out if there are any by simply eliminating them.
</span>

```python
{
    "tags": [
        "hide-input",
    ]
}

print(": Eliminating duplicates :", end='\n\n')

for disease, data in diseases.items():
    print(disease.ljust(15, ' '), end=' ')
    
    rows_before = data.shape[0]
    print(f'BEFORE({rows_before} rows)'.ljust(20, ' '), end=' ')
    
    data.drop_duplicates(inplace=True)
    
    rows_after = data.shape[0]
    print(f' AFTER({rows_after} rows)'.ljust(20, ' '), f'> {(rows_after/rows_before)*100:.2f}% of data remains.')
```

<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

Fortunately, there are no duplicates.

<h4 id="numerical-data"> |2.| Numerical Data Only </h4>
</span>
<br>
<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

During initial inspection of the files, `Cases` and `Incidence/100,000` were the only columns in each dataset that had numerical data. Let's take a quick look to make sure although the data is visually numerical they are stored as numbers and not as a text as it may happen sometimes while data is written into a file.

</span>

```python
{
    "tags": [
        "hide-input",
    ]
}

for disease, data in diseases.items():
    data.Year = data.Week.astype(np.int64)
    print(f"{disease.ljust(12, ' ')} dataset's numerical columns : {data.select_dtypes(include='number').columns.values}")
```

<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

Clearly, `polio` dataset stands out with `non-numerical` data type for its `Cases` data.

<br>Let's take a closer look this column to see if atleast its data is all numerical.
</span>

```python
{
    "tags": [
        "hide-input",
    ]
}

polio_bool = diseases['polio'][diseases['polio'].Cases.map(lambda c: not c.isdigit())]
polio_bool[['Cases', 'Incidence/100,000']].value_counts().to_frame().reset_index()
```

<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

**47** entries in `polio` have **non-numerical data**; `\N  Cases` and `0.0  Incidence/100,000`, which does not yield any meaningful information.

<br>I will **replace** `\N` with `0` for Cases.
</span>
<br>

```python
{
    "tags": [
        "hide-input",
    ]
}

print(f'polio: BEFORE Cases dtype = {diseases["polio"].Cases.dtypes}')

polio_cases = diseases['polio'].Cases.map(lambda c: 0 if c == '\\N' else int(c))
diseases['polio'].Cases = polio_cases

print(f'polio:  AFTER Cases dtype = {diseases["polio"].Cases.dtypes}')
```

<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

The **dtype = int64** (numerical data type) listed AFTER conversion means that the conversion was successful and it is stored as a number and not as text.

<h4 id="all-zeros"> |2.| All Zeros! </h4>
</span>
<br>
<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

First, let's take a look at how many 0's were reported for `hepatitis` dataset.
</span>

```python
{
    "tags": [
        "hide-input",
    ]
}

diseases['hepatitis'][
    diseases['hepatitis'][
        ['Cases', 'Incidence/100,000']
    ].apply(np.sum, axis=1) == 0].shape[0]
```

<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

`20114 rows of no data in hepatitis` alone!

<br>I can think of a couple possibilities for this scenario:
- No new disease incidence for the period.
- No cases were reported / they were missed.

<br>**No data, no value!** I wouldn't impute these data any further as [this article](https://www.nejm.org/doi/suppl/10.1056/NEJMms1215400/suppl_file/nejmms1215400_appendix.pdf) from which the disease data has been sourced has **already standardized data**. Anything 0, will be ignored. Next step is to **rid the data of 0's** in both `Cases` & `Incidence/100,000` columns in each of the datasets.
</span>
<br>

```python
{
    "tags": [
        "hide-input",
    ]
}

for disease in diseases:
    print(disease.ljust(15, ' '), end=' ')
    
    rows_before = diseases[disease].shape[0]
    print(f'BEFORE({rows_before} rows)'.ljust(20, ' '), end=' ')
    
    diseases[disease] = diseases[disease][
        diseases[disease][['Cases', 'Incidence/100,000']].apply(np.sum, axis=1) > 0]
    
    rows_after = diseases[disease].shape[0]
    print(f' AFTER({rows_after} rows)'.ljust(20, ' '), f'> {(rows_after/rows_before)*100:.2f}% of data remains.')
```

<h4 id="identify-season"> |2.| Identify the Season </h4>

</span>
<br>
<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align:justify;">

In addition, I'm interested in knowing if there's any yearly seasonality to the prevalence of the disease. For this, I will roughly classify the data into a `Season` based on the `Week` of the year:

|Season|Week of the year|
|:--|:--:|
|WINTER|Weeks 1-11, Weeks >= 52|
|SPRING|Weeks 12-25|
|SUMMER|Weeks 26-38|
|FALL|Weeks 39-51|

<br>In order to realize this, let's look at the `Week` column:
</span>

```python
{
    "tags": [
        "hide-input",
    ]
}

# Define Seasons
seasons = {52: 'WINTER', 39: 'FALL', 26: 'SUMMER', 12: 'SPRING', 1: 'WINTER'}

# Function to classify weeks into seasons
def what_season(week):
    for week_of_year in seasons.keys():
        if int(week) >= week_of_year:
            return seasons[week_of_year]
        
contagious_diseases[['Year', 'Week']] = (
    contagious_diseases.Week.apply(lambda wk: wk[:4] + ' ' + wk[-2:]).str.split(' ', expand=True)
)

contagious_diseases['Season'] = contagious_diseases.Week.map(what_season)
contagious_diseases.Season = pd.Categorical(contagious_diseases.Season, ['WINTER', 'SPRING', 'SUMMER', 'FALL'])

print('Categorized weeks into seasons [WINTER, SPRING, SUMMER, FALL]')

contagious_diseases.sample(4)
```

<h4 if="big-picture"> |F.| Seasonality in Disease Incidence </h4>
</span>
<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align: justify;">

<br>Seasonality is any predictable pattern that repeats timely. Although vaccination programs have aided in the decline of disease incidences, I am interested in a particular time of the year a disease is most likely at its peak. Let's draw focus on the peak times of the disease:
</span>
<br>

```python

cmap = mpl.colors.LinearSegmentedColormap.from_list("", ["ivory","firebrick","gold"])

sns.set_context('talk')
cd_dysc = contagious_diseases[['Disease', 'Year', 'Season', 'Incidence/100,000']]
cd_dysc_subset = cd_dysc.groupby(['Season', 'Year']).sum().reset_index()
cd_dysc_subset = cd_dysc_subset.pivot('Season', 'Year', 'Incidence/100,000')

plt.figure(figsize=(19,3))
ax = sns.heatmap(cd_dysc_subset, robust=True, cmap=cmap)

ax_cbar = ax.collections[0].colorbar
ax_cbar.ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

ax.tick_params(axis='y', which='major', labelcolor='lightgray', labelsize=20)
[y.set_color('gray') for y in ax.get_yticklabels()[:2]]
[y.set_fontweight('semibold') for y in ax.get_yticklabels()[:2]]
[y.set_size(25) for y in ax.get_yticklabels()[:2]]
ax.tick_params(axis='x', which='major', labelcolor='gray', labelsize=15)

plt.axhline(y=4, color='black', linewidth=1.5, alpha=0.7)
plt.xlabel('')
plt.ylabel('')

plt.savefig('contagiousdiseases-favorable-seasons.png', dpi=300, transparent=True)
plt.show()
```

<span style="font-family:candara, roboto, helvetica, courier, times; font-size:1.2em; font-style:italic; line-height: 145%; color:gray; text-align: justify;">

As you can see, the **disease incidence is the highest** in the `WINTER` & `SPRING` when the general immunity declines (perhaps due to minimal benefits of sunshine during this time of the year) and **lowest in** `SUMMER` & `FALL` (perhaps due to the highest benefits one can obtain due to optimal sunshine).
